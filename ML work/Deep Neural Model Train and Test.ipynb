```python
import pandas as pd
csv_path = "data/Data.csv"
df = pd.read_csv(csv_path, encoding="ISO-8859-1")
df.head()
```

```python
df.columns = ['season', 'age', 'diseases', 'accident', 'surgery', 'fever', 'alcohol', 'smoking', 'sitting', 'diagnosis']
```

```python
df_one = pd.get_dummies(df, columns=['fever'], prefix = ['fever'])
df_two = pd.get_dummies(df_one, columns=['alcohol'], prefix = ['alcohol'])
df_three = pd.get_dummies(df_two, columns=['smoking'], prefix = ['smoking'])
```

```python
df_three.loc[df_three['sitting'] > 0.5, 'sitting'] = 1
```

```python
df_three.loc[df_three['sitting'] <= 0.5, 'sitting'] = 0.5
```

```python
df_four = pd.get_dummies(df_three, columns=['sitting'], prefix = ['sitting'])
```

```python
df_five = pd.get_dummies(df_four, columns=['season'], prefix = ['season'])
```

```python
df_five.loc[df_five['age'] > 0.5, 'age'] = 1
```

```python
df_five.loc[df_five['age'] <= 0.5, 'age'] = 0.5
```

```python
df_six = pd.get_dummies(df_five, columns=['age'], prefix = ['age'])
df_six.head()
```

```python
import numpy as np
import tensorflow as ts

from sklearn.preprocessing import LabelEncoder

from keras.models import Sequential
from keras.utils import to_categorical
from keras.layers import Dense
```

```python
print(df_six.dtypes)
print(df_six.values)
```

```python
X = df_six.drop("diagnosis", axis=1)
y = df_six["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y = label_encoder.transform(y_train)
```

```python
for label, original_class in zip(encoded_y, y):
    print(original_class)
    print(label)
```

```python
from keras.utils import to_categorical
one_hot_y = to_categorical(encoded_y)
one_hot_y
```

```python
model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=22))
model.add(Dense(units=6, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, one_hot_y, epochs=300, shuffle=True, verbose=2)
```
