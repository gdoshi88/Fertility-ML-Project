```python
# Dependencies
import pandas as pd
```

## ETL: Data Extraction

```python
# Store filepath in a variable
file = "data/Data.csv"
```

```python
# Read our Data file with the pandas library
df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
# Show header
df.head()
```

```python
# Change column names
df.columns = ['season', 'age', 'diseases', 'accident', 'surgery', 'fever', 'alcohol', 'smoking', 'sitting', 'diagnosis']
```

```python
# Show df with new column names
df.head()
```

## ETL: Data Transformation

```python
#splitting fever column
df_one = pd.get_dummies(df, columns=['fever'], prefix = ['fever'])
df_one.head()
```

```python
#splitting alcohol column
df_two = pd.get_dummies(df_one, columns=['alcohol'], prefix = ['alcohol'])
df_two.head()
```

```python
#splitting smoking column
df_three = pd.get_dummies(df_two, columns=['smoking'], prefix = ['smoking'])
df_three.head()
```

```python
df_three.loc[df_three['sitting'] > 0.5, 'sitting'] = 1
```

```python
df_three.head()
```

```python
df_three.loc[df_three['sitting'] <= 0.5, 'sitting'] = 0.5
```

```python
df_three.head()
```

```python
df_four = pd.get_dummies(df_three, columns=['sitting'], prefix = ['sitting'])
```

```python
df_five = pd.get_dummies(df_four, columns=['season'], prefix = ['season'])
```

```python
df_five.loc[df_five['age'] > 0.5, 'age'] = 1
```

```python
df_five.loc[df_five['age'] <= 0.5, 'age'] = 0.5
```

```python
df_five.head()
```

```python
df_six = pd.get_dummies(df_five, columns=['age'], prefix = ['age'])
```

```python
df_six.head()
```

```python
df_six.columns
```

```python
#renaming columns:

df_six.rename(columns={
    'fever_-1': 'fever_1yr',
    'fever_0': 'fever_<3mo',
    'fever_1': 'fever_3mo_1yr',
    'alcohol_0.2': 'alcohol_mday',
    'alcohol_0.4': 'alcohol_1day',   
    'alcohol_0.6': 'alcohol_mweek',
    'alcohol_0.8': 'alcohol_1week',
    'alcohol_1.0': 'alcohol_negligent',
    'smoking_-1': 'smoke_never',
    'smoking_0': 'smoke_occ',
    'smoking_1': 'smoke_daily',
    'sitting_0.5': 'sitting_<=8hrs',
    'sitting_1.0': 'sitting_>8hrs',
    'season_-1.0': 'season_winter',
    'season_-0.33': 'season_spring',
    'season_0.33': 'season_summer',
    'season_1.0': 'season_fall',
    'age_0.5': 'age_18-27',
    'age_1.0': 'age_28-36'
}, inplace=True)

df_six.columns
```

```python
X = df_six.drop("diagnosis", axis=1)
y = df_six["diagnosis"]
print(X.shape, y.shape)
```

## Logistic Regression

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

```python
importances = rf.feature_importances_
importances
```

## Decision Tree

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

## SVC - Serial Vector Machines

```python
from sklearn.svm import SVC
```

```python
SVCmodel = SVC(kernel='linear')
SVCmodel
```

```python
SVCmodel.fit(X_train, y_train)

predictions = SVCmodel.predict(X_test)
predictions
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, predictions, target_names=['Normal', 'Abnormal']))

```
